```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
df = pd.read_csv("working_data.csv")
```

```python
df.head()
```

```python
x=df["Mean Distance (km)"]/1000
y=df["Gas"]

min_x = x.min()+1
min_y = y.min()+0.2
(slope, intercept,rvalue,pvalue,stderr)= linregress(x,y)
regress_values = x*slope + intercept
line_eq = "y = "+str(round(slope,2)) +"x + " + str(round(intercept,2))

plt.grid(True)
plt.xlabel('Average Distance (1000 km)')
plt.ylabel("Gas Price (USD)")
plt.plot(x,regress_values,"r-")
plt.scatter(x,y,marker='o')
plt.annotate(line_eq,(min_x,min_y),fontsize=15, color="red")
plt.title("Average Distance of Refineries")
print(f"The R-square = {round(rvalue**2,2)}")
plt.show()
```

```python
x=df["Min Distance (km)"]/1000
y=df["Gas"]

min_x = x.min()+0.5
min_y = y.min()+0.2
(slope, intercept,rvalue,pvalue,stderr)= linregress(x,y)
regress_values = x*slope + intercept
line_eq = "y = "+str(round(slope,2)) +"x + " + str(round(intercept,2))

plt.grid(True)
plt.xlabel('Minimum Distance (1000 km)')
plt.ylabel("Gas Price (USD)")
plt.plot(x,regress_values,"r-")
plt.scatter(x,y,marker='o')
plt.annotate(line_eq,(min_x,min_y),fontsize=15, color="red")
print(f"The R-square = {round(rvalue**2,2)}")
plt.show()
```
