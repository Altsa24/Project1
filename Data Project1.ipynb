```python
import requests

url = "https://gas-price.p.rapidapi.com/allUsaPrice"

headers = {
    'x-rapidapi-host': "gas-price.p.rapidapi.com",
    'x-rapidapi-key': "0fa0e84343msh23d24acd0946d22p14fce9jsn3487236e723f"
    }

response = requests.request("GET", url, headers=headers)

print(response.text)
```

```python
data= response.json()
```

```python
data["result"]
```

```python
states=[]
```

```python
gasoline=[]
midGrade=[]
premium=[]
diesel=[]
lista=data["result"]
for x in lista:
    states.append(x["name"])
    gasoline.append(x["gasoline"])
    midGrade.append(x["midGrade"])
    premium.append(x["premium"])
    diesel.append(x["diesel"])
    
```

```python
import pandas as pd 
gasoline_df = pd.DataFrame({"states":states,"gasoline":gasoline,"midGrade":midGrade,"premium":premium,"diesel":diesel})
gasoline_df.head()
```

```python
gasoline_df.to_csv("gasoline0406.csv",index=False)
```
